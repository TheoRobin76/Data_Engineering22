```
top_followers = ntj.sortBy(lambda x: x.user.followers_count, False)
```

```
ntj.filter(lambda x: x.user is None).take(3)
```

```
ntjf = ntj.filter(lambda x: 'user' in x)\
          .filter(lambda x: x.user is not None)\
          .filter(lambda x: 'followers_count' in x.user)\
          .filter(lambda x: x.user.followers_count is not None)
```

```
top_followers = ntjf.sortBy(lambda x: x.user.followers_count, False)
```

```
top_followers.first().asDict()
```

```
gamespot = top_followers.first()
gamespot.user.followers_count
gamespot.user.screen_name
gamespot.text
```

```
# Twitter Handles and follower counts for the top 20 tweeting accounts in terms of followers
top_followers.map(lambda x: x.user.screen_name).take(20)
```

```
handles_followers = ntjf.map(lambda x: (x.user.screen_name, x.user.followers_count))
handles_followers.take(5)
```

```
display(handles_followers.reduceByKey(max).sortBy(lambda x: x[1], False).take(20))
```

```
handles_followers.groupByKey().take(10)
```

```
handles_followers.groupByKey().mapValues(max).sortBy(lambda x: x[1], False).take(20)
```

```
# How many times did gamespot tweet?
ntjf.filter(lambda x: x.user.screen_name == "gamespot").count()
```

```
# Find the 10 users who tweeted the most during the livestream
ntjf.map(lambda x: (x.user.screen_name, 1)).reduceByKey(lambda x, y: x + y).sortBy(lambda x: x[1], False).take(10)
```

```
from wordcloud import WordCloud, STOPWORDS
```

```
text = ntjf.map(lambda x: x.text).collect()
textstr = " ".join(text)
```

```
STOPWORDS.update(["https", "NintentoE3", "RT", "t", "co", "E32018", "NintendoDirect", "NintendoE3"])
```

```
wc = WordCloud(stopwords=STOPWORDS).generate(textstr)
```

```
import matplotlib.pyplot as plt
```

```
x = plt.figure()
plt.imshow(wc)
plt.axis("off")
display(x)
```
