# JOINS

```
irisavg = iris.groupby('species').agg({'petal_length': 'avg'})
irismax = iris.groupby('species').agg({'petal_length': 'max'})
```

```
irisavg.show()
```

```
irismax.show()
```

```
irisavg.join(irismax, irisavg.species == irismax.species).show()
```

```
irisavg.join(irismax, "species").show()
```
